```python
import numpy as np
import pandas as pd
import statsrat as sr
from gradual_extn import spont_rec
from extn_rw_models import *
from plotnine import *
```

```python
#model = refined
#model = tdrva_decay
#model = tdrva_decay_cfg2
#model = CompAct_decay
model = revised_CompAct
save_plots = True
perform_oat = False
print(model.pars)

par_vals = {'refined': [0.001, 0.2, 1.5, 5.0],
            'tdrva_decay': [0.001, 0.25, 0.8, 1.0, 1.0, 5],
            'tdrva_decay_cfg2': [0.001, 0.25, 0.8, 1.0, 1.0, 5],
            'CompAct_decay': [0.01, 0.4, 2.0, 8.0, 5],
            'revised_CompAct': [0.01, 2.0, 0.2, 10.0, 0.9, 5]}
```

```python
if perform_oat:
    oat_result = sr.perform_oat(experiment = spont_rec,
                                model = model,
                                oat = 'grd_vs_std',
                                minimize = False,
                                n = 1,
                                max_time = 120)
    print(np.round(oat_result[0], 4))
    print(np.round(oat_result[1], 4))
```

```python
print('gradual extinction: spontaneous recovery \n')

trials = spont_rec.make_trials('gradual')
ds_grd = model.simulate(trials, par_vals[model.name])

trials = spont_rec.make_trials('standard')
ds_std = model.simulate(trials, par_vals[model.name])

trials = spont_rec.make_trials('gradual_reverse')
ds_rev = model.simulate(trials, par_vals[model.name])

print('\ngradual vs. standard')
print(spont_rec.oats['grd_vs_std'].compute_total({'gradual': ds_grd, 'standard': ds_std}))
print(spont_rec.oats['grd_vs_std'].mean_resp({'gradual': ds_grd, 'standard': ds_std}))
print('\ngradual vs. gradual reverse')
print(spont_rec.oats['grd_vs_rev'].compute_total({'gradual': ds_grd, 'gradual_reverse': ds_rev}))
print(spont_rec.oats['grd_vs_rev'].mean_resp({'gradual': ds_grd, 'gradual_reverse': ds_rev}))
```

```python
plot_b = sr.multi_plot(ds_list = [ds_grd, ds_std, ds_rev],
                       var = 'b',
                       rename_coords = {'schedule': 'group'},
                       only_main = True,
                       draw_points = True,
                       sel = 3*[{'t': ds_grd['stage_name'].isin(['cond', 'extinction', 'test'])}])
plot_b += theme(axis_text_x = element_text(ha = 'left'))
plot_b.draw()
if save_plots:
    plot_b.save(filename = 'plots/rw_gradual_spont_rec_b_' + model.name + '.jpeg', dpi = 300)
```

```python
plot_w = sr.multi_plot(ds_list = [ds_grd, ds_std, ds_rev],
                       var = 'w',
                       schedule_facet = True,
                       rename_coords = {'schedule': 'group', 'f_name': 'feature'},
                       text_size = 15,
                       stage_labels = False,
                       figure_size = (10, 3),
                       sel = 3*[{'f_name': ['cs', 'ctx'], 't': ds_grd['stage_name'].isin(['cond', 'extinction'])}])
plot_w += geom_hline(yintercept = 0, linetype = "dashed")
plot_w += scale_x_continuous(breaks = [0, 30], labels = ['cond', 'extinction'], name = 'stage')
plot_w += theme(axis_text_x = element_text(ha = 'left'))
plot_w.draw()
if save_plots:
    plot_w.save(filename = 'plots/rw_gradual_spont_rec_w_' + model.name + '.jpeg', dpi = 300)
```

```python
if 'atn' in ds_grd.keys():
    plot_eta = sr.multi_plot([ds_grd, ds_std, ds_rev],
                             var = 'atn',
                             rename_coords = {'f_name': 'feature'},
                             schedule_facet = True,
                             text_size = 15,
                             stage_labels = False,
                             figure_size = (10, 3),
                             sel = 3*[{'f_name': ['cs', 'ctx'], 't': ds_grd['stage_name'].isin(['cond', 'extinction'])}])
    plot_eta += ylab('eta')
    plot_eta += scale_x_continuous(breaks = [0, 30], labels = ['cond', 'extinction'], name = 'stage')
    plot_eta += theme(axis_text_x = element_text(ha = 'left'))
    plot_eta.draw()
    if save_plots:
        plot_eta.save(filename = 'plots/rw_gradual_spont_rec_eta_' + model.name + '.jpeg', dpi = 300)
```
