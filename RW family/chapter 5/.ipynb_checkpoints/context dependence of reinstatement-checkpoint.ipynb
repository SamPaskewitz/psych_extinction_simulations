```python
import numpy as np
import pandas as pd
import statsrat as sr
from reinstatement_context import reinstatement
from extn_rw_models import *
from plotnine import *
```

```python
model = refined
save_plots = True
perform_oat = False
print(model.pars)

par_vals = {'refined': [0.001, 0.2, 1.0, 5.0]}
```

```python
if perform_oat:
    oat_result = sr.perform_oat(experiment = reinstatement,
                                model = model,
                                oat = 'context_effect',
                                minimize = True,
                                n = 1)
    print(np.round(oat_result[0], 4))
    print(np.round(oat_result[1], 4))
```

```python
print('context dependence of reinstatement \n')

trials = reinstatement.make_trials('different')
ds_diff = model.simulate(trials, par_vals[model.name])

trials = reinstatement.make_trials('same')
ds_same = model.simulate(trials, par_vals[model.name])

print(reinstatement.oats['context_effect'].compute_total({'different': ds_diff, 'same': ds_same}))
print(reinstatement.oats['context_effect'].mean_resp({'different': ds_diff, 'same': ds_same}))
```

```python
plot_b = sr.multi_plot(ds_list = [ds_diff, ds_same],
                       var = 'b',
                       rename_coords = {'schedule': 'group'},
                       only_main = True,
                       draw_points = True,
                       sel = 2*[{'t': ds_diff['stage_name'].isin(['cond', 'extinction', 'test'])}])
plot_b += scale_color_manual(values = {'same': 'red', 'different': 'blue'})
plot_b.draw()
if save_plots:
    plot_b.save(filename = 'plots/rw_context_reinstatement_b_' + model.name + '.jpeg', dpi = 300)
```

```python
plot_w = sr.learn_plot(ds_same,
                       var = 'w',
                       rename_coords = {'f_name': 'feature'},
                       text_size = 15,
                       figure_size = (6, 4))
plot_w += geom_hline(yintercept = 0, linetype = "dashed")
plot_w.draw()
if save_plots:
    plot_w.save(filename = 'plots/rw_context_reinstatement_w_' + model.name + '.jpeg', dpi = 300)
```
