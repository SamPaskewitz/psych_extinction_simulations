```python
import numpy as np
import pandas as pd
import statsrat as sr
from statsrat.expr.predef.pvl_iti import aba_renewal, abc_renewal, aab_renewal
from extn_rw_models import *
from plotnine import *
```

```python
model = refined
save_plots = True
print(model.pars)

par_vals = {'refined': [0.001, 0.2, 1.5, 5.0]}
```

```python
print('comparing renewal designs \n')

trials = aba_renewal.make_trials('experimental')
ds_aba = model.simulate(trials, par_vals[model.name])
ds_aba.attrs['schedule'] = 'aba'

trials = abc_renewal.make_trials('experimental')
ds_abc = model.simulate(trials, par_vals[model.name])
ds_abc.attrs['schedule'] = 'abc'

trials = aab_renewal.make_trials('experimental')
ds_aab = model.simulate(trials, par_vals[model.name])
ds_aab.attrs['schedule'] = 'aab'
```

```python
plot_b = sr.multi_plot(ds_list = [ds_aba, ds_abc, ds_aab],
                       var = 'b',
                       rename_coords = {'schedule': 'group'},
                       only_main = True,
                       draw_points = True,
                       text_size = 15)
plot_b.draw()
if save_plots:
    plot_b.save(filename = 'plots/rw_renewal_designs_b_' + model.name + '.jpeg', dpi = 300)
```

```python
plot_w_aba = sr.learn_plot(ds_aba,
                           var = 'w',
                           rename_coords = {'f_name': 'feature'},
                           text_size = 15,
                           figure_size = (4, 4))
plot_w_aba += geom_hline(yintercept = 0, linetype = "dashed")
plot_w_aba.draw()
if save_plots:
    plot_w_aba.save(filename = 'plots/rw_renewal_designs_aba_w_' + model.name + '.jpeg', dpi = 300)
```

```python
plot_w_abc = sr.learn_plot(ds_abc,
                           var = 'w',
                           rename_coords = {'f_name': 'feature'},
                           text_size = 15)
plot_w_abc += geom_hline(yintercept = 0, linetype = "dashed")
plot_w_abc.draw()
if save_plots:
    plot_w_abc.save(filename = 'plots/rw_renewal_designs_abc_w_' + model.name + '.jpeg', dpi = 300)
```

```python
plot_w_aab = sr.learn_plot(ds_aab,
                           var = 'w',
                           rename_coords = {'f_name': 'feature'},
                           text_size = 15)
plot_w_aab += geom_hline(yintercept = 0, linetype = "dashed")
plot_w_aab.draw()
if save_plots:
    plot_w_aab.save(filename = 'plots/rw_renewal_designs_aab_w_' + model.name + '.jpeg', dpi = 300)
```
