```python
import sympy as sym
sym.init_printing(use_unicode=False, wrap_line=True)
```

```python
# https://www.sympy.org/en/index.html
x = sym.Symbol('x', real = True)
nu = sym.Symbol('nu', positive = True, real = True)
tau1 = sym.Symbol('tau1', real = True)
tau2 = sym.Symbol('tau2', positive = False, real = True)
sigma = sym.Symbol('sigma', positive = True, real = True)
```

```python
df = nu + 3
lhs0 = 0.5*((nu + 1)/(2*nu*(nu + 3)))**(-0.5)
lhs1 = ((2*nu)/(nu*tau2 - tau1**2))**(-0.5*df)
lhs2 = (0.5*(tau2 + x**2 - (tau1 + x)**2/(nu + 1)))**(-0.5*(df + 1))
lhs = lhs0*lhs1*lhs2
lhs
```

```python
mu = tau1/nu
t = (x - mu)/sigma
rhs = (1/sigma)*(1 + (1/df)*(t**2))**(-0.5*(df + 1))
rhs
```

```python
sym.solveset(lhs - rhs, sigma, domain = sym.Reals)
```
