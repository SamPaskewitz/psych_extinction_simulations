```python
import numpy as np
import pandas as pd
import statsrat as sr
from statsrat import latent_cause as lc
from statsrat.expr.predef.oc_renewal_no_iti import oc_renewal
from plotnine import *
```

```python
#model = lc.model(name = 'constant', kernel = lc.kernel.constant)
model = lc.model(name = 'power', kernel = lc.kernel.power)
print(model.pars)
save_plots = True
perform_oat = False

#method = 'particle'
method = 'local_vb'

prior_tau2_x = 1.0
prior_nu_x = 5.0
prior_tau2_y = 0.5
prior_nu_y = 2.0
alpha = 2.0
stick = 0.0
power = 1.0

par_vals = {'constant': [prior_tau2_x, prior_nu_x, prior_tau2_y, prior_nu_y, stick, alpha, 5.0],
            'power': [power, prior_tau2_x, prior_nu_x, prior_tau2_y, prior_nu_y, stick, alpha, 5.0]}
```

```python
print('occasion setting renewal \n')

trials_diff = oc_renewal.make_trials('different')
trials_same = oc_renewal.make_trials('same')

if method == 'particle':
    ds_diff = model.particle_filter(trials_diff, par_vals[model.name], n_p = 100)
    ds_same = model.particle_filter(trials_same, par_vals[model.name], n_p = 100)
    #sr.multi_plot([ds_diff, ds_same], var = 'b', only_main = True).draw()
    #sr.multi_plot([ds_diff, ds_same], var = 'mean_N').draw()
else:
    ds_diff = model.simulate(trials_diff, par_vals[model.name])
    ds_same = model.simulate(trials_same, par_vals[model.name])
    #sr.multi_plot([ds_diff, ds_same], var = 'b', only_main = True).draw()
    #sr.multi_plot([ds_diff, ds_same], var = 'phi', sel = 2*[{'z_name': ['0', '1']}], schedule_facet = True).draw()
    #sr.multi_plot([ds_diff, ds_same], var = 'N').draw()
    df = ds_diff['z'].to_dataframe()
    table = pd.crosstab(df['z'], df['ex'])
    print(ds_diff['z'].values)
    print(table)
    
plot_b = sr.multi_plot([ds_diff, ds_same],
                        var = 'b',
                        draw_points = True,
                        dodge_width = 0.2,
                        rename_coords = {'schedule': 'group'})
plot_b += scale_color_manual(values = {'different': 'red', 'same': 'blue'})
plot_b.draw()
if save_plots:
    plot_b.save(filename = 'plots/latent_cause_oc_renewal_b_' + model.name + '.jpeg', dpi = 500)

oat_name = list(oc_renewal.oats.keys())[0]
print(oc_renewal.oats[oat_name].compute_total({'different': ds_diff, 'same': ds_same}))
print(oc_renewal.oats[oat_name].mean_resp({'different': ds_diff, 'same': ds_same}))
```

```python
if perform_oat:
    n_dict = {'local_vb': 1, 'particle': 10}
    oat_result = sr.perform_oat(model = model,
                                experiment = oc_renewal,
                                n = n_dict[method],
                                minimize = False,
                                sim_type = method,
                                max_time = 200)
    print(np.round(oat_result[0], 4))
    print(np.round(oat_result[1], 4))
```
